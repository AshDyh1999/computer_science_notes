### Regular Expressions
$L(r)$: language(set of strings) generated by the regular expression
+ character set that is avaiable
+ set elements are referred as symbols
+ $r$ is a regular expression

$\Sigma$ = alphabet: set of legal symbols

In a regular expression, all symbols indicate *patterns*.

A regular expression $r$ may contain characters that have special meanings. Such characters are referred as *metacharacters* or *metasymbols*. An *escape character* "turns off" their special meaning.
Escape characters themselves are metacharacters, if they are also legal in the alphabet.

#### Basic Regular Expressions
$L(\mathbf{a}) = \{a\}$: regular expression **a** matches the character *a*

$L(\epsilon) = \{\epsilon\}$: a match of a string that is empty

$L(\phi) = \{\}$: matches no string
#### Operations

+ Choice among alternatives(lowest precendence): $L(r|s) = L(r) \cup L(s)$
+ Concatenation: $L(rs) = L(r)L(s)$: The first part of the string matches **r**, while the second part matches **s**. Example: $L((\mathbf{a}|\mathbf{b})\mathbf{c}) = L(\mathbf{a}|\mathbf{b})L(\mathbf{c}) = \{a, b\}\{c\} = \{ac, bc\}$
+ Repetition(highest precedence): $L(r*) = L(r)*$. Example 1: **a\*** matches *a*, *aa*, *aaa* and so on; Example 2: $S^* = \{\epsilon\} \cup S \cup SS \cup SSS \cup \dots$


#### Names
+ ***digit*** = 0|1|2|...|9
+ ***digit\**** = (0|1|2|...|9)*

`X = Y * 0` is the same as `X = 0` if `X` is integer, but invalid if `X` is a float number because of the existence of `NaN`. `NaN * 0 = NaN`, and thus `X` does not equal to 0.
